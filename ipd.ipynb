```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
date_columns = ['start_dates', 'end_dates', 'issue_dates', 'expiry_dates']
for col in date_columns:
    resume_data[col] = pd.to_datetime(resume_data[col], errors='coerce')
imputer = SimpleImputer(strategy='mean')
resume_data['matched_score'] = imputer.fit_transform(resume_data[['matched_score']])
resume_data.columns = resume_data.columns.str.replace('ï»¿', '')
if 'job_position_name' in resume_data.columns:
    resume_data.drop(columns=['job_position_name'], inplace=True)
else:
    print("Column 'job_position_name' not found in DataFrame.")
resume_data.head()
```

```python
numeric_df = resume_data.select_dtypes(include=[np.number])
if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(10, 8))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
sns.histplot(resume_data['matched_score'], kde=True)
plt.title('Distribution of Matched Scores')
plt.xlabel('Matched Score')
plt.ylabel('Frequency')
plt.show()
```
